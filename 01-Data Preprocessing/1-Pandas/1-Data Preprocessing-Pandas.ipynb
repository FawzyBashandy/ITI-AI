### Null-Values Handling

```python
my_dict2 = {
    'Name':['alice','Bob','Charlie','Abdelrahman','youssef','fawzy',None],
    'Age':[30,35,20,27,23,None,23],
    'City':['New York','Los Angeles',None,'cairo','Giza','cairo','cairo']
}
df2 = pd.DataFrame(my_dict2)
df2
```

### Detect Null Values

```python
df2.isnull()
```

```python
df2.isnull().sum() # More interpretable
```

### Fill Null Values

```python
df2['City'].fillna(" ",inplace=False) # Fill with empty String
```

```python
df2['City'].fillna(df['City'].mode()[0],inplace=False) # Fill with empty String
## mode() returns a list of most repeated value so we take first element example [cairo=5,alexanderia=5] take anyone we will take cairo iindex=0
```

```python
print(df2['Age'])
df2['Age'].fillna(int(df2['Age'].mean()),inplace=False)
```

### Drop Null Values

```python
df2.dropna(inplace=False) #Drop any row that has a null value in it
```

```python
df2['Age'].dropna(inplace=False) # Drop only rows that has "Age' is NULL
```

### all the previous i was doing inplace = False ( Because i don't want original df to be affected)

```python
df2.dropna(inplace=True)
print(df2.isnull().sum())
print("\n")
print(df2.info())
```

## Groupby() group rows based on a condition

```python
my_dict3 = {
    'Name':['alice','Bob','Charlie','Abdelrahman','youssef','fawzy','Seif'],
    'Age':[30,35,20,27,23,24,23],
    'City':['New York','Los Angeles','Chicago','Giza','Giza','cairo','cairo'],
    'Salary':[2000,3000,4000,5000,6000,7000,8000]
}
df3 = pd.DataFrame(my_dict3)
df3
```

```python
df3.groupby(by='City')['Age'].mean() #Group by city then for each group calculate mean of age
```

## Apply Function for a Colomn (Feature)

```python
def double_salary(salary):
    return salary*2

df3['Double_Salary'] = df3['Salary'].apply(double_salary) # Create new colomn by applying function to existing colomn
df3
```

### Apply Lambda Function

```python
df3['Double_SalaryLambda'] = df3['Salary'].apply(lambda x: x*2) # Using lambda function to do the same
df3
```
