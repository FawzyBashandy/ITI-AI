### Avg of list

```python
def avg(list_):
    sum=0
    count = 0
    for i in list_:
        sum += i
        count +=1
    return sum/count
avg([1,2,3])
```

### ✅ Better & Pythonic way (recommended)

```python
def avg(lst):
    return sum(lst) / len(lst)

avg([1,2,3])
```

### func that returns the tallest string in list of strings

```python
def tallest_string(strings):
    if not strings:
        return None, None

    max_idx = 0
    max_str = strings[0]

    for i, s in enumerate(strings):
        if len(s) > len(max_str):
            max_str = s
            max_idx = i

    return max_idx, max_str

tallest_string(["Faw","F"])
```

### Pythonic way

```python
def tallest_string(strings):
    if not strings:
        return None, None
    idx, value = max(enumerate(strings), key=lambda x: len(x[1]))
    return idx, value
```

### Func that returns prime numbers inside a list

```python
def getPrime(numlist):
    primeList=[]
    for x in numlist:
        isPrime = True
        for dev in range(2,x):
            if x % dev == 0:
                isPrime = False
                break
        if isPrime == True:
            primeList.append(x)
    return primeList
getPrime([2,3,4,5,6])
```

```python
class Product:
    name = 'default'
    price = 0.0
    quantity = 0

    def __init__(self,name,price):
        self.name = name
        self.price = price
    def restock(self,quantity):
        self.quantity += quantity
    def sell(self,quantity):
        if self.quantity < quantity:
            print("Stock Unavilable")
            return
        else:
            self.quantity -= quantity
    def display(self):
        print(f"Produce Name is : {self.name}\nPrice is:{self.price}\nAvailable Quantity is:{self.quantity}")

```

```python
product = Product("Milk",20)
product.display()
product.restock(40)
product.display()
product.sell(20)
product.display()
product.sell(40)
```

```python
class Calculator:
    def add(self, *numbers):
        total = 0
        for n in numbers:
            total += n
        return total

    def subtract(self, *numbers):
        if not numbers:
            return 0
        result = numbers[0]
        for n in numbers[1:]:
            result -= n
        return result

    def multiply(self, *numbers):
        result = 1
        for n in numbers:
            result *= n
        return result

    def divide(self, *numbers):

        result = numbers[0]
        for n in numbers[1:]:
            if n == 0:
                print("Division by zero Problem")
                break
            result /= n
        return result
```

```python
x= Calculator()
x.multiply(1,2,3)
```
