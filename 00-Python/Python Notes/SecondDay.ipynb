### Get Prime Numbers

```python
for x in range(1,30):
    if x>1:
        for i in range(2,int(x**0.5)+1):
            if x%i == 0:
                break
        
        else:
            print(f"{x} is prime")
        
```

### Count Vowels in string

```python
String = "hellO"
vowelsCounter = 0
for c in String:
    if c.lower() in "aeoui":
        vowelsCounter += 1
    
vowelsCounter
```

### Simulate your ATM

```python
balance = 1000
withdrawAmmount = 0
while balance:
    print(f'your current balance is {balance}')
    withdrawAmmount = int(input("Enter Withdraw Ammount"))
    if(withdrawAmmount <= balance):
        balance-=withdrawAmmount
    else:
        print("insufficent balance \n",f"your current balance is{balance}")
else:
    print(f'your balance has been consumed')
```

### Factorial

```python
number = int(input("Enter the Number: "))
factorial = 1
if number > 0:
    for i in range(number,0,-1): # 0 is excluded don't worry
        factorial *= i
    else:
        print(f"Factorial of {number} is {factorial}")
```

### Fibbonaci Series

```python
Fn0 =0
Fn1 =1
number = int(input("Enter the number : "))
if number > 2:
    for i in range(1,number):
        Fn = Fn1 + Fn0
        Fn0 = Fn1
        Fn1 = Fn
elif number == 1:
    Fn = Fn1
elif number == 0:
    Fn = Fn0
print(f"Fibbonaci of {number} is: {Fn}")
```

### Check Palindrome Strings

```python
string = input("Enter a string:")
for i in range(0,len(String)//2):
    if string[i].lower()== string[-i-1].lower(): # in python string[0] = string[-1] -> string[1] = string[-2]
        continue
    else:
        break
else:
    print(f'{string} is palindrome')
```

#### Checking Palindrome (Python Style)

```python
string = input("Enter a String:")
isPalindrome = False
if string.lower() == string[::-1].lower(): # string[::-1] returns the string Reversed
    isPalindrome =True
isPalindrome
```

### Merge two lists(key and value) into Dictionary

```python
names = ['ahmed','fawzy','mohamed']
ages = [28,24,22]
my_dict = {}
for name in range(0,len(ages)):
    actualname = names[name]
    my_dict[actualname] = ages[name]
my_dict
```

### do it again

```python
keys =["name","age",'grade','courses']
values =['fawzy','24','A-',['c','c++','linux admin']]
my_dict = {}
for i in range(0,len(keys)):
    my_dict[keys[i]]=values[i]
my_dict
```
